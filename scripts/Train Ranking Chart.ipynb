```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Goutham\DA PROJECTS\railway-performance-analytics\data\railway_data_2024.csv")

rank_df = df.groupby("Train_Name")["Delay_Minutes"].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
rank_df.head(20).plot(kind="bar")
plt.title("Top 20 Trains Ranked by Average Delay")
plt.xlabel("Train Name")
plt.ylabel("Average Delay (Minutes)")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig(r"C:\Users\Goutham\DA PROJECTS\railway-performance-analytics\outputs\train_ranking.png")

print("Train ranking chart created successfully!")
```
