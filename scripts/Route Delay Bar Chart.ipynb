```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Goutham\DA PROJECTS\railway-performance-analytics\data\railway_data_2024.csv")

# Create Route Column 
df["Route"] = df["Source"] + "  " + df["Destination"]

route_df = df.groupby("Route")["Delay_Minutes"].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
route_df.plot(kind="bar")
plt.title("Average Delay by Route")
plt.xlabel("Route")
plt.ylabel("Average Delay (Minutes)")
plt.xticks(rotation=90)
plt.tight_layout()

plt.savefig(r"C:\Users\Goutham\DA PROJECTS\railway-performance-analytics\outputs\route_delay_bar.png")

print("Route delay bar chart created successfully!")
```
