```python
import pandas as pd
```

```python
df = pd.read_csv("railway_data_2024.csv")
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(15)
```

```python
df['Delay_Category'] = df['Delay_Minutes'].apply(
    lambda x: 'On Time' if x==0 else
              'Low Delay' if x<=15 else
              'Medium Delay' if x<=45 else
              'High Delay'
)
```

```python
df.sample(10)
```

```python
monthly_delay = df.groupby('Month')['Delay_Minutes'].mean()
monthly_delay.plot(kind='line')
```

```python
monthly_rev = df.groupby('Month')['Revenue'].sum()
monthly_rev.plot(kind='bar')
```

```python
df.groupby('Date')['Passenger_Count'].sum().plot(kind='line')
```

```python
df.groupby("Train_Name")['Delay_Minutes'].mean().sort_values()
```

```python
df.to_csv("cleaned_dataset.csv")
```
